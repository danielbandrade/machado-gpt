```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import torch
from google.colab import drive
drive.mount('/content/drive')
```

```python
directory_path = '/content/drive/MyDrive/Pessoal @ UniAng/Programing /machado_txt'

# Ensure the directory exists to avoid errors
if not os.path.isdir(directory_path):
    print(f"Error: Directory not found at {directory_path}")
else:
    print(f"Directory '{directory_path}' found.")
    files_in_directory = os.listdir(directory_path)
    if not files_in_directory:
        print("No files found in the directory.")
    else:
        print("Top-level items found in the directory:")
        for f in files_in_directory:
            print(f"- {f}")

    concatenated_content = []
    txt_files_found = 0
    for root, _, files in os.walk(directory_path):
        for filename in files:
            if filename.endswith('.txt'):
                file_path = os.path.join(root, filename)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        concatenated_content.append(f.read())
                        txt_files_found += 1
                except Exception as e:
                    print(f"Error reading file {file_path}: {e}")

    # Join all content with a separator (e.g., a newline or double newline)
    final_text = "\n\n---\n\n".join(concatenated_content)

    # Print the concatenated content (or save it to a new file)
    if final_text:
        print(f"\nConcatenated content from {txt_files_found} .txt files:")
        #print(final_text)
    else:
        print("\nNo .txt files were found or processed to concatenate any content within the directory and its subdirectories.")
```

```python
print("lenght of dataset in caracters: ", len(final_text))
```

```python
# unique caracters
chars = sorted(list(set(final_text)))
vocab_size = len(chars)
print(''.join(chars))
print('\n')
print(vocab_size)
```

```python
char_freq = []
for char in chars:
  char_freq.append(final_text.count(char))

char_freq_dict = dict(zip(chars, char_freq))
```

```python
# Sort by values (descending order)
sorted_by_values_desc = dict(sorted(char_freq_dict.items(), key=lambda item: item[1], reverse=True))
print(sorted_by_values_desc)
```

```python
stoi = { ch:i for i,ch in enumerate(chars) }
itos = { i:ch for i,ch in enumerate(chars) }
encode = lambda s: [stoi[c] for c in s] # encoder: take a string, output a list of integers
decode = lambda l: ''.join([itos[i] for i in l]) # decoder: take a list of integers, output a string
```

```python
print(encode("olá, estou aqui!"))
print(decode(encode("olá, estou aqui!")))
```

```python
data = torch.tensor(encode(final_text), dtype=torch.long)
print(data.shape, data.dtype)
print(data[:100])
```
